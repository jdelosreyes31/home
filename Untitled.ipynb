```python
import pandas as pd
```

Load Kenpom

```python
pd.set_option('display.max_columns', 25)
df_html = pd.read_html('https://kenpom.com/')
```

```python
df = df_html[0]
df.columns = df.columns.get_level_values(1)
```

Get league average Tempo

```python
df_small = df.iloc[:,[1,2,4,5,7,9,11,15,17]].copy()
```

```python
df_small_fix = df_small.drop(
    [
        df_small.index[40],
        df_small.index[41],
        df_small.index[82],
        df_small.index[83],
        df_small.index[124],
        df_small.index[125],
        df_small.index[166],
        df_small.index[167],
        df_small.index[208],
        df_small.index[209],
        df_small.index[250],
        df_small.index[251],
        df_small.index[292],
        df_small.index[293],
        df_small.index[334],
        df_small.index[335],
    ]
)
```

```python
df_small_fix['AdjT'] = pd.to_numeric(df_small_fix['AdjT'])
df_small_fix['AdjO'] = pd.to_numeric(df_small_fix['AdjO'])
df_small_fix['AdjD'] = pd.to_numeric(df_small_fix['AdjD'])
df_small_fix['AdjEM'] = pd.to_numeric(df_small_fix['AdjEM'])
df_small_fix['Luck'] = pd.to_numeric(df_small_fix['Luck'])
df_small_fix['OppO'] = pd.to_numeric(df_small_fix['OppO'])
df_small_fix['OppD'] = pd.to_numeric(df_small_fix['OppD'])
```

```python
laT = df_small_fix['AdjT'].mean()
laT
```

Get league average points per posession

```python
Oppp = df_small_fix['AdjO'].mean()
Dppp = df_small_fix['AdjD'].mean()
ppp = ((Oppp * Dppp) / 100)
print(ppp)
```

GAMEDAY CALCULATION

```python
away_team = "North Carolina"
home_team = "NC State"

df_small_fix[df_small_fix['Team'].isin([home_team, away_team])]
```

```python
# Get HOME TEAM FIXTURES
hadjO = df_small_fix[df_small_fix['Team'] == home_team]['AdjO'].values[0]
hadjD = df_small_fix[df_small_fix['Team'] == home_team]['AdjD'].values[0]
hadjT = df_small_fix[df_small_fix['Team'] == home_team]['AdjT'].values[0]
hluck = df_small_fix[df_small_fix['Team'] == home_team]['Luck'].values[0]
hOppO = df_small_fix[df_small_fix['Team'] == home_team]['OppO'].values[0]
hOppD = df_small_fix[df_small_fix['Team'] == home_team]['OppD'].values[0]

# Get AWAY TEAM Fixtures
aadjO = df_small_fix[df_small_fix['Team'] == away_team]['AdjO'].values[0]
aadjD = df_small_fix[df_small_fix['Team'] == away_team]['AdjD'].values[0]
aadjT = df_small_fix[df_small_fix['Team'] == away_team]['AdjT'].values[0]
aluck = df_small_fix[df_small_fix['Team'] == away_team]['Luck'].values[0]
aOppO = df_small_fix[df_small_fix['Team'] == away_team]['OppO'].values[0]
aOppD = df_small_fix[df_small_fix['Team'] == away_team]['OppD'].values[0]
```

```python
#Predicted pace
xp = (hadjT * aadjT) / laT

#Clown Fiesta
if aadjD > 102 and aadjT > 73:
    adef = 6
else:
    adef = 0

# adjusted effiency away + home effiencies
aps = aadjO - (aadjO * (.000 + (aluck /10)))
apa = aadjD + (aadjD *(.000)) + adef

hps = (hadjO + (hadjO * (.000 + (hluck / 10) )))
hpa = hadjD - (hadjD *(.000))

print(xp)
```

```python
# home project points scored (IONA) adjusted to league average points per possesion
lgadjOa = hadjO / ((aOppO * aOppD) / ppp)
xhps = hps * lgadjOa

# multiply expected home score by expected away points allowed over league average points per possession
hproj = (hadjO * apa) / ppp
hprojT = round(((hproj * xp) / 100),0)

print("home adjusted points scored per 100 posessions", xhps, "adjusted away points allowed", apa)
print("non-adjusted prediction", hprojT)

if xp < 66:
    hprojT = hprojT - 8
    print("low end prediction", home_team, int(hprojT))
elif xp > 70 and hadjO > 105 and hluck > .100:
    hprojT = hprojT + 8
    print("medium end projection", home_team, hprojT)
elif (xp > 73 and hadjO > 105) or hluck > .125:
    hprojT = hprojT + 10
    print("high end prediction", home_team, int(hprojT))
else:
    print(home_team, hprojT)
```

```python
# Away projected points scored (MORG) adjusted to league average points per possession
lgadjOh = aadjO / ((hOppO * hOppD) / ppp)
xaps = aps * lgadjOh

# multiply expected home score by expected away points allowed over league average points per possession
aproj = (xaps * hpa) / ppp
aprojT = round(((aproj * xp) / 100), 0)

print("away adjusted points scored per 100 possessions", xaps, "adjusted home points allowed per 100 possessions", hpa)
print("non-adjusted points", aprojT)

if hadjD < 92:
    if hadjD < 88 and xp < 68:
        print(away_team, int(aprojT) - 3)
        aprojT = aprojT - 6
    else:
        print(away_team, int(aprojT) - 3)
        aprojT = aprojT - 3
elif xp > 70 and hadjD > 105:
    print(away_team, int(aprojT) + 3)
    aprojT = aprojT + 3
else:
    print(away_team, aprojT)
    
```

```python
#home spread
spread = hprojT - aprojT
#projected total
total = hprojT + aprojT

print(home_team, "by ", spread)
print("Total: ", total)
#print("Adjusted Total plus/minus:", xtotal)
```
